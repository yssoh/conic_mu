## SOCP Lifts of Regular 5gon

```python
import numpy as np
from scipy.linalg import circulant
import conicmu
```

```python
# Generate regular n-gon Slack Matrix
def GenerateSCngon(n):
    
    v = np.zeros((n,))
    for i in range(n):
        v[i,] = np.cos(np.pi/n) - np.cos((2*i-1)*np.pi/n)
    SC = circulant(v)

    return SC
```

```python
# Construct the Slack Matrix
SC = GenerateSCngon(5)
SC
```

```python
Profile = [(1,'socp')]
Output_1 = conicmu.factorize_parallel(SC,Profile,fname="reg5gon_1")
```
